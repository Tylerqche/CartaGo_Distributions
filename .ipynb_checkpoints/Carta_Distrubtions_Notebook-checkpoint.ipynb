# Carta Go Clean Data Distributions

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CARTA GO clean data file
df = pd.read_excel('CARTA GO clean data.xlsx')
```

## Data Distributions for Requested Pickup Time

Distributions are categorized in to their time components and also their weekday or weekend status.

Times can be changed to account of rush hours and busy conditions.

Currents Constraints:
Mornings 5am - 12pm,
Afternoons 12pm - 5pm,
Evenings 5pm - 9pm,
Nights 9pm - 5am,

```python
# Parse the date and time strings into datetime objects
df['Requested Pickup Time'] = pd.to_datetime(df['Requested Pickup Time'])

morning = (5, 12)
afternoon = (12, 17)
evening = (17, 21)
night = (21, 5)

def categorize_time_of_day(time):
    if time.hour >= morning[0] and time.hour < morning[1]:
        return 'Morning'
    elif time.hour >= afternoon[0] and time.hour < afternoon[1]:
        return 'Afternoon'
    elif time.hour >= evening[0] and time.hour < evening[1]:
        return 'Evening'
    else:
        return 'Night' # Schedules do not occuring during night
    
# Add a new column for the time of day category
df['Time of Day'] = df['Requested Pickup Time'].apply(categorize_time_of_day)

# Group the data by weekdays and weekends
df_weekday_request = df[df['Requested Pickup Time'].dt.weekday < 5]
df_weekend_request = df[df['Requested Pickup Time'].dt.weekday >= 5]

# Filter weekdays based on time of day 
weekday_morning_rows = df_weekday_request[df_weekday_request['Time of Day'] == 'Morning']
weekday_afternoon_rows = df_weekday_request[df_weekday_request['Time of Day'] == 'Afternoon']
weekday_evening_rows = df_weekday_request[df_weekday_request['Time of Day'] == 'Evening']

# Filter weekend based on time of day 
weekend_morning_rows = df_weekend_request[df_weekend_request['Time of Day'] == 'Morning']
weekend_afternoon_rows = df_weekend_request[df_weekend_request['Time of Day'] == 'Afternoon']
weekend_evening_rows = df_weekend_request[df_weekend_request['Time of Day'] == 'Evening']
```

### Time of Day vs. Number of Requests

Vastly more requests happen during weekdays rather than weekends.

```python
# Count the number of requests for each time of day category on weekdays
morning_count_weekday = weekday_morning_rows.shape[0]
afternoon_count_weekday = weekday_afternoon_rows.shape[0]
evening_count_weekday = weekday_evening_rows.shape[0]

# Count the number of requests for each time of day category on weekends
morning_count_weekend = weekend_morning_rows.shape[0]
afternoon_count_weekend = weekend_afternoon_rows.shape[0]
evening_count_weekend = weekend_evening_rows.shape[0]

# Define the width of the bars
bar_width = 0.35

# Define the x locations for the groups
x_weekday = np.arange(3)
x_weekend = x_weekday + bar_width

# Create a bar graph for weekday data
plt.bar(x_weekday, [morning_count_weekday, afternoon_count_weekday, evening_count_weekday], color='navy', width=bar_width, label='Weekdays')
# Create a bar graph for weekend data
plt.bar(x_weekend, [morning_count_weekend, afternoon_count_weekend, evening_count_weekend], color='forestgreen', width=bar_width, label='Weekends')

# Display the exact number of requests on top of each bar
for i, v in enumerate([morning_count_weekday, afternoon_count_weekday, evening_count_weekday]):
    plt.text(i, v + 10, str(v), fontsize=10, color='navy', ha='center')
for i, v in enumerate([morning_count_weekend, afternoon_count_weekend, evening_count_weekend]):
    plt.text(i + bar_width, v + 10, str(v), fontsize=10, color='forestgreen', ha='center')

plt.xlabel('Time of Day')
plt.ylabel('Number of Requests')
plt.title('Requests by Time of Day (Weekdays vs Weekends)')
plt.xticks(x_weekday + bar_width/2, ['Morning', 'Afternoon', 'Evening'])
plt.legend()

plt.show()
```

### Total Mean and Medium of Weekday and Weekend Requests

Weekdays exhibit considerably higher demand compared to weekends. On average, there are 1009.60 requests per week on weekdays, with a median of 999.00 requests. In contrast, weekends have much lower demand, with an average of 141.60 requests per week and a median of 142.00 requests. This data underscores the substantial disparity in service usage between weekdays and weekends.

```python
# Resample the DataFrame to get the number of requests per week
requests_per_week = df.resample('W', on='Requested Pickup Time').size()

# Resample the DataFrame for weekdays and weekends separately to get the number of requests per week
requests_per_weekend = df_weekend_request.resample('W', on='Requested Pickup Time').size()
requests_per_weekday = df_weekday_request.resample('W', on='Requested Pickup Time').size()

# Calculate the mean and median number of requests per week
mean_requests_per_week = requests_per_week.mean()
median_requests_per_week = requests_per_week.median()

print("Mean and Medium Requests for Weekends and Weekdays:")

# Print the mean and median requests per week
print("Mean requests per week: {:.2f}".format(mean_requests_per_week))
print("Median requests per week: {:.2f}".format(median_requests_per_week))

# Calculate the mean and median number of requests per week for weekends and weekdays separately
mean_requests_per_weekend = requests_per_weekend.mean()
median_requests_per_weekend = requests_per_weekend.median()
mean_requests_per_weekday = requests_per_weekday.mean()
median_requests_per_weekday = requests_per_weekday.median()

# Print the mean and median requests per week for weekends and weekdays separately
print("Mean requests per week (weekends): {:.2f}".format(mean_requests_per_weekend))
print("Median requests per week (weekends): {:.2f}".format(median_requests_per_weekend))
print("Mean requests per week (weekdays): {:.2f}".format(mean_requests_per_weekday))
print("Median requests per week (weekdays): {:.2f}".format(median_requests_per_weekday))
```

### Average Distance between Orign and Destination Locations

```python
# Calculate the differences in longitude and latitude
delta_lng = df['Destination Lng'] - df['Origin Lng']
delta_lat = df['Destination Lat'] - df['Origin Lat']

# Function to calculate distance
def calculate_distance(row):
    delta_lng = row['Destination Lng'] - row['Origin Lng']
    delta_lat = row['Destination Lat'] - row['Origin Lat']
    return np.sqrt(delta_lng**2 + delta_lat**2)

# Apply the calculate_distance function to each row to calculate the distance
distance = df.apply(calculate_distance, axis=1)

# Calculate the average distance
average_distance = distance.mean()

print("Average Distance between Origin and Destination:")
print("Longitude: {:.6f} degrees".format(delta_lng.mean()))
print("Latitude: {:.6f} degrees".format(delta_lat.mean()))
print("Euclidean Distance: {:.6f}".format(average_distance))
```

```python
# Calculate average distance for weekday mornings
average_distance_weekday_morning = weekday_morning_rows.apply(calculate_distance, axis=1).mean()

# Calculate average distance for weekday afternoons
average_distance_weekday_afternoon = weekday_afternoon_rows.apply(calculate_distance, axis=1).mean()

# Calculate average distance for weekday evenings
average_distance_weekday_evening = weekday_evening_rows.apply(calculate_distance, axis=1).mean()

# Calculate the differences in longitude and latitude for weekday mornings
delta_lng_weekday_morning = weekday_morning_rows['Destination Lng'] - weekday_morning_rows['Origin Lng']
delta_lat_weekday_morning = weekday_morning_rows['Destination Lat'] - weekday_morning_rows['Origin Lat']

# Calculate the differences in longitude and latitude for weekday afternoons
delta_lng_weekday_afternoon = weekday_afternoon_rows['Destination Lng'] - weekday_afternoon_rows['Origin Lng']
delta_lat_weekday_afternoon = weekday_afternoon_rows['Destination Lat'] - weekday_afternoon_rows['Origin Lat']

# Calculate the differences in longitude and latitude for weekday evenings
delta_lng_weekday_evening = weekday_evening_rows['Destination Lng'] - weekday_evening_rows['Origin Lng']
delta_lat_weekday_evening = weekday_evening_rows['Destination Lat'] - weekday_evening_rows['Origin Lat']

print("Average Euclidean Distance between Origin and Destination for Weekdays:")
print("Weekday Morning:")
print("  Longitude: {:.6f} degrees".format(delta_lng_weekday_morning.mean()))
print("  Latitude: {:.6f} degrees".format(delta_lat_weekday_morning.mean()))
print("  Euclidean Distance: {:.6f} units".format(average_distance_weekday_morning))

print("Weekday Afternoon:")
print("  Longitude: {:.6f} degrees".format(delta_lng_weekday_afternoon.mean()))
print("  Latitude: {:.6f} degrees".format(delta_lat_weekday_afternoon.mean()))
print("  Euclidean Distance: {:.6f} units".format(average_distance_weekday_afternoon))

print("Weekday Evening:")
print("  Longitude: {:.6f} degrees".format(delta_lng_weekday_evening.mean()))
print("  Latitude: {:.6f} degrees".format(delta_lat_weekday_evening.mean()))
print("  Euclidean Distance: {:.6f} units".format(average_distance_weekday_evening))
```

### Average Difference in Longitude and Latitude per Time of Day for each Day

```python
ordered_time = ['Morning', 'Afternoon', 'Evening']

# Group by day of the week and time of day
average_lng_per_day_time_of_day = df.groupby([df['Requested Pickup Time'].dt.day_name(), df['Time of Day']])['Delta Lng'].mean().reindex(ordered_time, level=1)

# Reindex the levels of MultiIndex to reorder days of the week
average_lng_per_day_time_of_day = average_lng_per_day_time_of_day.reindex(ordered_days, level=0)

# Print the average difference in longitude traveled for each day of the week and time of day
print("Average Difference in Longitude Traveled per Day of the Week and Time of Day:")
print(average_lng_per_day_time_of_day)
```

```python
# Group by day of the week and time of day
average_lat_per_day_time_of_day = df.groupby([df['Requested Pickup Time'].dt.day_name(), df['Time of Day']])['Delta Lat'].mean().reindex(ordered_time, level=1)

# Reindex the levels of MultiIndex to reorder days of the week
average_lat_per_day_time_of_day = average_lat_per_day_time_of_day.reindex(ordered_days, level=0)

# Print the average difference in latitude traveled for each day of the week and time of day
print("\nAverage Difference in Latitude Traveled per Day of the Week and Time of Day:")
print(average_lat_per_day_time_of_day)
```
